```python
import pandas as pd
rating_url = "ratings.csv"
rating_df = pd.read_csv(rating_url)

rating_df.head()
```

```python
import numpy as np
import math
from sklearn.metrics.pairwise import cosine_similarity
```

```python
rating_sparse_df = rating_df.pivot(index='user', columns='item',
                                   values='rating').fillna(0).reset_index().rename_axis(index=None, columns=None)
rating_sparse_df = rating_sparse_df.set_index('user')
rating_sparse_df.head()
```

```python
train_set = rating_sparse_df.iloc[0:1000,:]
train_set
```

```python
train_set.loc[:20]
```

```python
user_similarity = cosine_similarity(train_set)
user_similarity.shape
```

```python
user_similarity
```

```python
user_sim_df = pd.DataFrame(user_similarity,index=train_set.index,columns = train_set.index)
```

```python
user_sim_df
```

```python
def get_similar_user(user_name):
  similar_user = user_sim_df[user_name]
  similar_user = similar_user.sort_values(ascending=False)
  return similar_user[1:3]

sim_2 = get_similar_user(2)
sim_2
```
