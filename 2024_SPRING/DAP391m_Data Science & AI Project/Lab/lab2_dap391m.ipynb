Data Visualization with Python:

- A survey was conducted to gauge audience interest in different data science topics, namely:

    - Big Data (Spark / Hadoop)

    - Data Analysis Statistics

    - Data Journalism

    - Data Visualization

    - Deep Learning

    - Machine Learning

- The participants had three options for each topic:

    - Very Interested

    - Somewhat interested

    - Not interested.

- Total: 2,233 respondents completed the survey.

- Use the artist layer of Matplotlib to replicate the bar chart below to visualize the percentage of the respondents' interest in the different data science topics surveyed.

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_csv('/home/minhvu/Documents/minimal_ds/dap391m/lab2 Topic_Survey_Assignment.csv')

display(df.head())
df.describe()
```

### Use the artist layer of Matplotlib to replicate the bar chart below to visualize the percentage of the respondents' interest in the different data science topics surveyed.

```python
topics = df["Unnamed: 0"].tolist()
very_interested = df["Very interested"].tolist()
somewhat_interested = df["Somewhat interested"].tolist()
not_interested = df["Not interested"].tolist()
```

```python
total_participants = sum(very_interested) + sum(somewhat_interested) + sum(not_interested)
percentages = {
    topic: [
        (count / total_participants) * 100
        for count in (vi, si, ni)
    ]
    for topic, vi, si, ni in zip(topics, very_interested, somewhat_interested, not_interested)
}


fig, axes = plt.subplots(2, 3, figsize=(12, 8))

for i, (topic, values) in enumerate(percentages.items()):
    row, col = divmod(i, 3)
    wedge1, wedge2, wedge3 = axes[row, col].pie(values, labels=["Very interested", "Somewhat interested", "Not interested"], autopct="%1.1f%%", startangle=140)

    
    axes[row, col].set_title(topic)
    axes[row, col].axis("equal") 



plt.tight_layout()
plt.show()
```
